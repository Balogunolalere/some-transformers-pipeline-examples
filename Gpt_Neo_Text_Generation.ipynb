```python
from transformers import pipeline #import transformer
```

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-1.3B')
```

```python
prompt = 'to our friends in the west, our oppressors and tyrants' 
```

```python
res = generator(prompt, max_length=200, do_sample=True, temprature=0.9)
```
